Importing transformers packages with pipeline and Gradio.

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
import gradio as gr
```

Importing my pre-trained model and tokenizer from Google Drive and making classifier object with these entities.

```python
model=AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/NN_LLM_Project_Files')
tokenizer=AutoTokenizer.from_pretrained('/content/drive/MyDrive/NN_LLM_Project_Files')
classifier=pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

Making my own function that returns a score out of 5 and running a sample sentence with it.

```python
def review(dic):
  label=dic[0]['label']
  stars=0
  if label == 'LABEL_0':
    stars=1
  if label == 'LABEL_1':
    stars=2
  if label == 'LABEL_2':
    stars=3
  if label == 'LABEL_3':
    stars=4
  if label == 'LABEL_4':
    stars=5
  return f"{stars} out of 5 star review"

def output_message(line):
  return review(classifier("This is a bad place."))

#running an example
print(review(classifier("This is a fine place.")))
output_message("This is not that good. I don't like it.")
```

Making Gradio Interface and running it.

```python
iface = gr.Interface(
    fn=output_message,
    inputs="text",  # The input is text (comment) provided by the user
    outputs="text",  # The output is text (the rating)
    title="Customer Review Rating",  # Title of the app
    description="Enter a comment and check what's the rating given by the customer."  # Description of the app
)

if __name__ == "__main__":
    iface.launch(share=True)
```
